### 2.5.- Manipulando LinkExtractor - Extracción con Scrapy en FarmaciaVerde

```python
from scrapy.item import Field
from scrapy.item import Item
from scrapy.spiders import CrawlSpider, Rule
from scrapy.selector import Selector
from scrapy.linkextractors import LinkExtractor
from scrapy.loader import ItemLoader
```

```python
class Farmacia(Item):
    Nombre=Field()
    Precio=Field()

class CruzVerde(CrawlSpider):
    name = 'Farmacias'
    custom_settings = {
        'USER-AGENT': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)Chrome/80.0.3987.149 Safari/537.36',
        'CLOSESPIDER_PAGECOUNT':30
    }
 
    allowed_domains = ["cruzverde.cl"]
 
    start_urls = ["https://www.cruzverde.cl/medicamentos/"]
    rules = (
        Rule(LinkExtractor(
            allow=r'start=', 
            tags=('a', 'button'), # tags en los que LinkExtractor debe buscar
            attrs=('href', 'data-url') # atributos dentro de los tags en los que LinkExtractor debe buscar
        ), follow=True, callback="parse_farmacia"),
    )
    download_delay = 1
    
    def parse_farmacia(self,response):
        sel = Selector(response)
        productos = sel.xpath('//div[@class="col-12 col-lg-4"]')
        
        for x in productos:
            item = ItemLoader(Farmacia(), x)
 
            item.add_xpath('Nombre', 
                           './/div[@class="tile-body px-3 pt-3 pb-0 d-flex flex-column pb-0"]//div[@class="pdp-link"]/a/text()')
 
            item.add_xpath('Precio',
                           './/span[contains(@class, "value ")]/text()')
 
            yield item.load_item()
```

```python
from scrapy.spiders import Spider
from scrapy.crawler import CrawlerProcess

if __name__ == "__main__": # Código que se va a ejecutar al dar clic en RUN
        process = CrawlerProcess()
        process.crawl(CruzVerde) # Nombre de la clase de mi Spider
        process.start()
```
